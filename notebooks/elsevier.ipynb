```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_table('/Users/jaredlorince/Desktop/parse_log',header=None,names=['uid','abstract','formatted','raw','wc_raw','wc_parsed'])
```

```python
raw = df[df.raw==1]
formatted = df[df.formatted==1]
abs_only = df[(df.abstract==1)&(df.formatted==0)&(df.raw==0)]
```

```python
bins = np.arange(0,10000,100)
```

```python
plt.plot(bins[:-1],np.histogram(raw.wc_parsed,bins=bins)[0],label='raw text')
plt.plot(bins[:-1],np.histogram(formatted.wc_parsed,bins=bins)[0],label='formatted text')
plt.plot(bins[:-1],np.histogram(abs_only.wc_parsed,bins=bins)[0],label='abstracts_only')

plt.plot(bins[:-1],np.histogram(raw.wc_raw,bins=bins)[0],ls='--',c='blue')
plt.plot(bins[:-1],np.histogram(formatted.wc_raw,bins=bins)[0],ls='--',c='green')
plt.plot(bins[:-1],np.histogram(abs_only.wc_raw,bins=bins)[0],ls='--',c='red')

plt.legend()
plt.xlabel("Word_count (solid=parsed, dashed=raw)")
plt.ylabel('Number of papers')
#plt.xscale('log')


```
